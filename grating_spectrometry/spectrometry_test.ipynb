```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def diffraction(k, phi, n, g):
    return 2 *np.cos(k) *np.sin(phi) /g /n
```

```python
a = np.deg2rad(33.9)
b = np.arange(-np.pi/2, np.pi/2, np.pi/200)

ks = a
phis = b -a

wavelengths = diffraction(ks, phis, 1, 300000)
```

```python
ax = plt.subplot(1,1,1)

im = plt.imshow([wavelengths], cmap= "rainbow", aspect= "auto")
plt.colorbar(im)
plt.show()
```
